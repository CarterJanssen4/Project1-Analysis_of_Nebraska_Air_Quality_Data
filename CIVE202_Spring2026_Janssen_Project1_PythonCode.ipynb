## Formatting Data

```python
## Load data
import pandas as pd

air_data = pd.read_csv('AirQuality_Daily_StudentVersion.csv')

## Turn data into a dataframe
air_data = pd.DataFrame(air_data)

(air_data)
```

```python


#Pivot Table

pivot_stats = pd.pivot_table(
    air_data,
    index = 'sensor.name',
    values = ['pm2.5_atm', 'pm10.0_atm', 'voc'],
    aggfunc = ['mean', 'median'])

pivot_stats
```

## 1) 5 Locations with the highest mean and median values for pm1.5_atm, pm1.0_atm, and voc

#### pm2.5_atm

```python
air_data_pm25 = air_data.groupby(['sensor.name']).agg(
                                                     mean = ('pm2.5_atm', 'mean'), 
                                                     median = ('pm2.5_atm', 'median'))



top5_pm25 = air_data_pm25.sort_values(by = ['mean','median'], ascending = False)
top5_pm25.head(5)

```

#### pm10.0_atm

```python
air_data_pm10 = air_data.groupby(['sensor.name']).agg(
                                                     mean = ('pm10.0_atm', 'mean'), 
                                                     median = ('pm10.0_atm', 'median'))



top5_pm10 = air_data_pm10.sort_values(by = ['mean','median'], ascending = False)
top5_pm10.head(5)
```

#### VOC

```python
air_data_voc = air_data.groupby(['sensor.name']).agg(
                                                     mean = ('voc', 'mean'), 
                                                     median = ('voc', 'median'))



top5_voc = air_data_voc.sort_values(by = ['mean','median'], ascending = False)
top5_voc.head(5)
```

## 2) On what days did the maximum values occur and where did they occur. 

#### Location

```python

#Group Data
air_data_pm25 = air_data.groupby(['sensor.name']).agg(max = ('pm2.5_atm', 'max'))


#Sort Data
top5_pm25 = air_data_pm25.sort_values(by = 'max', ascending = False)

top5_pm25.head(10)

```

#### Date

```python

#Group Data
air_data_pm25 = air_data.groupby(['date']).agg(max = ('pm2.5_atm', 'max'))


#Sort Data
top5_pm25 = air_data_pm25.sort_values(by = 'max', ascending = False)

top5_pm25.head(10)
```

## 3) Does humidity and Temperature have an effect on air quality

```python
#Humidity
#Define functions

def assign_humidity(X) :
    if X < 50:
        return 'Low'
    elif 50 <= X <= 80:
        return 'High'
    elif X > 80:
        return 'Very High'
    else:
        return 'No Data'

# Apply your function
air_data['humidity_level'] = air_data['humidity'].apply(assign_humidity)

air_data.head(10)
```

```python
#Temperature

#Define functions
def assign_temperature(X) :
    if X < 32:
        return 'Below Freezing'
    elif 32 <= X <= 50:
        return 'Cool'
    elif 50 < X <= 70:
        return 'Warm'
    elif X > 70:
        return 'Hot'
    else:
        return 'No Data' 

# Apply your function
air_data['temperature_level'] = air_data['temperature'].apply(assign_temperature)

air_data.head(10)
```

## 4) Air Quality Index

#### PM 2.5

```python
#Define Function
def assign_air_quality(X) :
    if X > 35.5:
        return 'health risk'
    elif X <= 35.5:
        return 'healthy' 
    else:
        return 'No Data'

#Assign Function
air_data['air_quality_pm2.5'] = air_data['pm2.5_atm'].apply(assign_air_quality)

#Sort by highest pm2.5 value
top5_pm25 = air_data.sort_values ('pm2.5_atm', ascending = False)

top5_pm25.head(50)
```

#### PM 10.0

```python
#Define Function
def assign_air_quality(X) :
    if X > 155:
        return 'health risk'
    elif X <= 155:
        return 'healthy' 
    else:
        return 'No Data'

#Assign Function
air_data['air_quality pm10.0'] = air_data['pm10.0_atm'].apply(assign_air_quality)

#Sort by highest pm10.0 value
top5_pm10 = air_data.sort_values ('pm10.0_atm', ascending = False)

top5_pm10.head(50)
```

### Unhealthy Dates

```python
### Apply filters

#Dates with highest VOC, and locations
#Only care about data that has pm2.5 > 35.5

unhealthy_pm25 = air_data[air_data['pm2.5_atm'] >= 35.5 ] 

unhealthy_dates = unhealthy_pm25.sort_values('date')



unhealthy_dates = unhealthy_pm25[['date','sensor.name','pm2.5_atm']].sort_values('date')

unhealthy_dates
```

```python
### Apply filters

#Dates with highest VOC, and locations
#Only care about data that has pm10.0 > 155

unhealthy_pm10 = air_data[air_data['pm10.0_atm'] >= 155 ] 

unhealthy_dates = unhealthy_pm10.sort_values('date')



unhealthy_dates = unhealthy_pm10[['date','sensor.name','pm10.0_atm']].sort_values('date')

unhealthy_dates.head(10)
```
